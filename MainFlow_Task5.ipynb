```python
# 5. Scatter plot to visualize the distribution based on age and cholesterol
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='chol', hue='target', palette='coolwarm', data=df)
plt.title('Distribution of Heart Disease Based on Age and Cholesterol Levels')
plt.xlabel('Age')
plt.ylabel('Cholesterol Level (mg/dL)')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
#6. How Does the Slope of the Peak Exercise ST Segment (Slope) Correlate with the Presence of Heart Disease?
# Plot the relationship between slope and target
sns.boxplot(x='target', y='slope', data=df)
plt.title('Correlation between Slope of Peak Exercise ST Segment and Heart Disease')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Slope of Peak Exercise ST Segment')
plt.show()

# Calculate average slope for patients with and without heart disease
slope_heart_disease = df[df['target'] == 1]['slope']
slope_no_heart_disease = df[df['target'] == 0]['slope']

print(f"Average Slope with Heart Disease: {slope_heart_disease.mean()}")
print(f"Average Slope without Heart Disease: {slope_no_heart_disease.mean()}")
```

```python
#7. Plot the relationship between number of major vessels (ca) and heart disease
sns.countplot(x='ca', hue='target', data=df)
plt.title('Trend between Number of Major Vessels (ca) and Heart Disease')
plt.xlabel('Number of Major Vessels (ca)')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()

# Calculate average number of vessels (ca) for patients with and without heart disease
ca_heart_disease = df[df['target'] == 1]['ca']
ca_no_heart_disease = df[df['target'] == 0]['ca']

print(f"Average Number of Major Vessels (ca) with Heart Disease: {ca_heart_disease.mean()}")
print(f"Average Number of Major Vessels (ca) without Heart Disease: {ca_no_heart_disease.mean()}")
```
