```python
columns=data1.columns
print(columns)
```

Finding number of missing values in each column and it's percentage

```python
# Calculate the percentage of missing values for each column
missing_values = data1[columns].isnull().sum()
total_rows = len(data1)
missing_values_percentage = (missing_values / total_rows) * 100
```

```python
# Create a summary DataFrame for missing values
missing_summary = pd.DataFrame({
    'Missing Values': missing_values,
    'Missing Values Percentage': missing_values_percentage
})
```

```python
# Print the summary table for missing values
print("Summary of Missing Values:\n", missing_summary)
```

```python
# Identify numeric columns
numeric_cols = data3.select_dtypes(include=['float64']).columns

# Calculate Q1, Q3, and IQR for each numeric column
Q1 = data3[numeric_cols].quantile(0.25)
Q3 = data3[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers for numeric columns only
data3 = data3[~((data2[numeric_cols] < (Q1 - 1.5 * IQR)) | (data3[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

print("DataFrame after removing outliers from numeric columns:\n", data3)
```
